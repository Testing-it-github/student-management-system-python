```python
###### import csv
import os

#create csv file====================================================================================

FILE_NAME = "students.csv"

def create_file():
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["ID", "Name", "Course", "Mobile"])
        print("csv file created successfully")

create_file()
        
#Add student detail=====================================================================================

def add_student():

    sid = input("Enter Student ID: ")

    # check digits in ID
    if not sid.isdigit():
        print("Student ID must contain only digits")
        return

    # check duplicate ID
    with open(FILE_NAME, "r") as f:
        reader = csv.reader(f)
        for row in reader:
            if row[0] == sid:
                print("Student ID already exists!")
                return

    name = input("Enter Name: ")

    if not name.replace(" ", "").isalpha():       
        print("Name must contain only alphabets")
        return

    course = input("Enter Course: ")

    if not course.replace(" ", "").isalpha():
        print("Course must contain only alphabets")
        return

    mobile = input("Enter Mobile Number: ")

    if not mobile.isdigit() or len(mobile) != 10:
        print("Enter valid 10 digit mobile number")
        return

    with open(FILE_NAME, "a", newline="") as f:
        writer = csv.writer(f)
        writer.writerow([sid, name, course, mobile])

    print("Student added successfully!")

#view student============================================================================================

def view_students():
    with open(FILE_NAME, "r") as f:
        reader = csv.reader(f)

        next(reader)  

        rows = list(reader)  

        if not rows:
            print("No student records found.")  
        else:
            print("\nID\tName\tCourse\tMobile") 
            for row in rows:
                print("\t".join(row))

#search student================================================================================================

def search_student():
    search = input("enter student ID or Name to search: ")

    found = False

    with open(FILE_NAME, "r") as f:
        reader = csv.reader(f)
        next(reader)

        for row in reader:
            if search == row[0] or search.lower() == row[1].lower():                
                print("\nStudent Found:")
                print("ID:", row[0])
                print("Name:", row[1])
                print("Course:", row[2])
                print("Mobile:", row[3])
                found = True
                break
            
    if not found:
            print("Student not Found")

#update student====================================================================================================

def update_student():
    sid = input("Enter Student ID to update: ")
    updated_rows = []
    found = False

    with open(FILE_NAME, "r") as f:
        reader = csv.reader(f)
        next(reader)

        for row in reader:
            if row[0] == sid:
                print("Student Found. Enter new details.")

                name = input("Enter New Name: ")
                course = input("Enter New Course: ")
                mobile = input("Enter New Mobile: ")

                updated_rows.append([sid, name, course, mobile])
                found = True
            else:
                updated_rows.append(row)

    if found:
        with open(FILE_NAME, "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["ID", "Name", "Course", "Mobile"])  # ← add this line
            writer.writerows(updated_rows)
        print("Student updated successfully!")
    else:
        print("Student ID not found")

#delete student details==================================================================================================

def delete_student():
    sid = input("Enter Student ID to delete: ")
    rows = []
    found = False

    with open(FILE_NAME, "r") as f:
        reader = csv.reader(f)
        next(reader)

        for row in reader:
            if row[0] == sid:
                found = True
                print("Student record deleted")
            else:
                rows.append(row)

    if found:
        with open(FILE_NAME, "w", newline="") as f:
            writer = csv.writer(f)

            writer.writerow(["ID","Name","Course","Mobile"])  
            writer.writerows(rows)
    else:
        print("Student ID not found")


while True:
    print("\n====== Student Management System ======")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        view_students()

    elif choice == "3":
        search_student()

    elif choice == "4":
        update_student()

    elif choice == "5":
        delete_student()

    elif choice == "6":
        print("exit")
        break

    else:
        print("Invalid choice")
    
```
